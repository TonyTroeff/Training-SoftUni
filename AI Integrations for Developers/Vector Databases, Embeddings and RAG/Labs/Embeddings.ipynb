```python
import numpy as np

from pprint import pprint
from rich.console import Console
from rich.table import Table
from typing import List

def cosine_similarity(e1: List[float], e2: List[float]) -> np.float64:
    v1 = np.array(e1)
    v2 = np.array(e2)

    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

## Vectors

$$
\vec{a} = (a_1, a_2, \dots, a_n)
$$

## What is the `dot product` of two vectors?

### Algebraic definition

$$
\vec{a} \cdot \vec{b} = \sum_{i=1}^n{a_i b_i} \\
$$

The length (also called magnitude or Euclidean norm) of a vector is defined as:
$$
\|\vec{a}\| = \sqrt{\vec{a} \cdot \vec{a}} = \sqrt{\sum_{i=1}^n{a_i^2}}
$$

### Geometric definition

From the geometric form, we can derive cosine similarity, which measures the similarity between two vectors, considering only their orientation.

$$
\vec{a} \cdot \vec{b} = \|\vec{a}\| \cdot \|\vec{b}\| \cdot \cos(\theta) \\
\therefore \cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{\|\vec{a}\| \cdot \|\vec{b}\|}
$$

```python
from google.colab import userdata
from openai import OpenAI

api_key = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=api_key)
```

```python
phrases = ["a cat is on the mat", "it is a beautiful day", "it is sunny outside", "garfield loves lasagna", "math is hard", "math is sooo easy"]
```

```python
create_embeddings_response = client.embeddings.create(
    model="text-embedding-3-small",
    input=phrases
)
```

```python
embeddings = [el.embedding for el in create_embeddings_response.data]
```

```python
console = Console()

table = Table(show_header=True)
table.add_column("Phrase #1")
table.add_column("Phrase #2")
table.add_column("Similarity", justify="right")

for i in range(len(phrases)):
    for j in range(i + 1, len(phrases)):
        similarity = cosine_similarity(embeddings[i], embeddings[j])
        table.add_row(phrases[i], phrases[j], f"{similarity:.5f}")


console.print(table)
```
