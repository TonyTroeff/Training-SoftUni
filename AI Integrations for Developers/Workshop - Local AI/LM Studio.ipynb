```python
from openai import OpenAI

client = OpenAI(base_url="http://localhost:1234/v1", api_key="not_used")
```

```python
def print_response(response):
    print(f"Response id: {response.id}")
    print(f"Input tokens: {response.usage.input_tokens} ({response.usage.input_tokens_details.cached_tokens} cached); Output tokens: {response.usage.output_tokens} ({response.usage.output_tokens_details.reasoning_tokens} reasoning)")

    for response_item in response.output:
        print()

        if response_item.type == "reasoning":
            print(f"\033[93m{'-' * 20} [Thoughts] {'-' * 20}\033[00m")
            for content_item in response_item.content:
                print(f"\033[93m{content_item.text}\033[00m")
        elif response_item.type == "message":
            print(f"{'-' * 20} [Output] {'-' * 20}")
            for content_item in response_item.content:
                print(content_item.text)
```

```python
available_models = client.models.list()
```

```python
for model in available_models.data:
    print(model)
```

```python
simple_response = client.responses.create(
    model="nvidia/nemotron-3-nano-4b",
    input="Find the flaw in this argument: All smart people drink coffee. I drink coffee. Therefore I’m smart."
)
```

```python
print_response(simple_response)
```
