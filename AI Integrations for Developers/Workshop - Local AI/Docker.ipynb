```python
from openai import OpenAI
from openai.types.chat import ChatCompletion, ChatCompletionUserMessageParam

client = OpenAI(base_url="http://localhost:12434/v1", api_key="not_used")
```

```python
def print_chat_completion(completion: ChatCompletion):
    print(f"Completion id: {completion.id}")

    input_tokens = completion.usage.prompt_tokens
    cached_tokens = 0 if completion.usage.prompt_tokens_details is None else completion.usage.prompt_tokens_details.cached_tokens
    output_tokens = completion.usage.completion_tokens
    reasoning_tokens = 0 if completion.usage.completion_tokens_details is None else completion.usage.completion_tokens_details.reasoning_tokens
    print(f"Input tokens: {input_tokens} ({cached_tokens} cached); Output tokens: {output_tokens} ({reasoning_tokens} reasoning)")

    if len(completion.choices) > 1:
        raise Exception("Expected a single choice.")

    message = completion.choices[0].message
    if message is None:
        raise Exception("Expected a single message.")

    print(message.content)
```

```python
available_models = client.models.list()
```

```python
for model in available_models.data:
    print(model)
```

```python
simple_response = client.chat.completions.create(
    model="docker.io/ai/smollm3:Q8_0",
    messages=[
        ChatCompletionUserMessageParam(role="user", content="Find the flaw in this argument: All smart people drink coffee. I drink coffee. Therefore I’m smart.")
    ]
)
```

```python
print_chat_completion(simple_response)
```
