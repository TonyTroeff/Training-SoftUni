```python
import base64

from IPython.display import Image, Markdown
from pathlib import Path

def print_images_response(response):
    if response.usage is not None:
        print(f"Input tokens: {response.usage.input_tokens}; Output tokens: {response.usage.output_tokens}")

    for image in response.data:
        if image.revised_prompt is not None:
            display(Markdown(image.revised_prompt))

        image_bytes = base64.b64decode(image.b64_json)
        display(Image(data=image_bytes, format=response.output_format))

def save_image(response, path_to_file: Path):
    if len(response.data) != 1:
        raise Exception("Expected response to contain a single image")

    image_bytes = base64.b64decode(response.data[0].b64_json)
    with path_to_file.open(mode="wb") as file:
        file.write(image_bytes)
```

```python
from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
generate_portrait_prompt = """
                           A photorealistic portrait of a software developer. He is 24-years-old, from Italy, and wears glasses.
                           The scene is illuminated by the sunlight streaming through the open window, creating a welcoming and calm atmosphere.
                           Through the window a beautiful garden can be seen with all kinds of fruit trees and vegetables.
                           The overall mood is serene and masterful. Vertical portrait orientation.
                           """
```

## Generate an image using `GPT image models`

```python
generate_portrait_response = client.images.generate(
    model="gpt-image-1-mini",
    prompt=generate_portrait_prompt,
    size="1024x1024",
    quality="low",
    output_format="png"
)
```

```python
print_images_response(generate_portrait_response)
```

```python
save_image(generate_portrait_response, Path("/content/portrait-dalle.png"))
```

## Generate an image using `Dall-E models`

```python
generate_portrait_response = client.images.generate(
    model="dall-e-3",
    prompt=generate_portrait_prompt,
    size="1024x1024",
    response_format="b64_json",
    quality="standard"
)
```

```python
print_images_response(generate_portrait_response)
```

```python
save_image(generate_portrait_response, Path("/content/portrait-dalle.png"))
```
