```python
from IPython.display import Image, Markdown, Video

def print_generation(generation):
    display(Markdown("### Thumbnail"))
    display(Image(url=generation.assets.image, embed=True))
    display(Markdown("### Video"))
    display(Video(generation.assets.video, embed=True))
```

```python
from google.colab import userdata
from lumaai import LumaAI

api_key = userdata.get('LUMAAI_API_KEY')
client = LumaAI(auth_token=api_key)
```

```python
import time

def generate(*args, **kwargs):
    generation = client.generations.create(*args, **kwargs)

    for i in range(1000):
        if generation.state == "completed":
            return generation
        if generation.state == "failed":
            raise RuntimeError(f"Generation failed: {generation.failure_reason}")

        time.sleep(2)
        generation = client.generations.get(id=generation.id)

    raise RuntimeError("Video generation couldn't finish in time.")
```

## Text to Video

```python
funny_cat_prompt = "Generate a short and funny video of a cat that is playing with its tail, while laying on a mat."
```

```python
funny_cat_generation = generate(
    prompt=funny_cat_prompt,
    model="ray-flash-2",
    resolution="540p",
    duration="5s"
)
```

```python
print_generation(funny_cat_generation)
```

## Image to Video

```python
walking_cat_prompt = "Extend this schene with a cat walking on the kitchen counter."
```

```python
walking_cat_generation = generate(
    prompt=walking_cat_prompt,
    model="ray-flash-2",
    resolution="540p",
    duration="5s",
    keyframes={
        "frame0": {
            "type": "image",
            "url": "https://images.pexels.com/photos/2089698/pexels-photo-2089698.jpeg"
        }
    }
)
```

```python
print_generation(walking_cat_generation)
```
