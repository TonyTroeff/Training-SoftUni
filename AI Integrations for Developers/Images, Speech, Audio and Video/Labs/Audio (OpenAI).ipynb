```python
from IPython.display import Audio
from pathlib import Path
from typing import Optional

def print_transcript(transcript):
    print(f"Input tokens: {transcript.usage.input_tokens} ({transcript.usage.input_token_details.audio_tokens} audio); Output tokens: {transcript.usage.output_tokens}")

    print()
    print(f"{'-' * 20} [Text] {'-' * 20}")
    print(transcript.text)

def display_audio(path_to_file: Path):
    display(Audio(path_to_file))
```

```python
from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

## Text to Speech

```python
article_text = """
               It is also important to consider the bigger picture — our test infrastructure, and more precisely, the combinatorial setup, should be applicable in the current situation and any future derivatives as well.
               As mentioned several times already, we are looking for a future-proof solution — one that can withstand any further development of the system, whether planned or yet unforeseen.
               For example, we may need to introduce a new set of components responsible for managing subscriptions — we should be able to incorporate them within our tests along with all payment methods and processors.
               """
```

```python
article_speech = client.audio.speech.create(
    input=article_text,
    model="gpt-4o-mini-tts",
    voice="onyx"
)
```

```python
PATH_TO_ARTICLE_AUDIO = Path("/content/article.mp3")
article_speech.write_to_file(PATH_TO_ARTICLE_AUDIO)
```

```python
display_audio(PATH_TO_ARTICLE_AUDIO)
```

## Speech to Text

```python
def transcribe(path_to_file: Path, language: Optional[str] = None):
    with path_to_file.open(mode="rb") as file:
        transcript = client.audio.transcriptions.create(
            model="gpt-4o-mini-transcribe",
            file=file,
            language=language,

            # NOTE: Tony Troeff, 16/03/2026 - This prompt is optional.
            prompt="The recording is in Bulgarian and may mention SoftUni, students, and career center activities."
        )

    return transcript
```

```python
PATH_TO_CAREER_CENTER_AUDIO = Path("/content/career_center_softuni.mp3")
display_audio(PATH_TO_CAREER_CENTER_AUDIO)
```

```python
career_center_transcript = transcribe(PATH_TO_CAREER_CENTER_AUDIO, language="bg")
```

```python
print_transcript(career_center_transcript)
```
