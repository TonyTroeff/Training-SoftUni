```python
from google.colab import userdata
from langchain_community.document_loaders import GithubFileLoader, GitHubIssuesLoader
```

```python
github_token = userdata.get('GITHUB_TOKEN')

github_file_loader = GithubFileLoader(
    repo="TonyTroeff/Training-SoftUni",
    access_token=github_token,
    file_filter=lambda path: path.find('#') == -1 and path.endswith(".md")
)
```

```python
github_file_loader.load()
```

```python
github_issues_loader = GitHubIssuesLoader(
    repo="TryAtSoftware/CleanTests",
    access_token=github_token,
    per_page=50
)
```

```python
github_issues_loader.load()
```
